## List of Queries to be performed with their time implementation

* **Query 1 : Displaying all the data from total energy production and total energy consumption with SELECT and WHERE Query**
* **Query 2 : Using SUM & AVG as calculation function along with Select and Order to get an Average Consumption per Country**
* **Query 3 : Using GROUP BY and Comparison in the WHERE clause to select those countries which have consumption rate greater than the production rate in the year 1990**
* **Query 4 : Applying ORDER BY queries to get the list in ascending/descending order**
* **Query 5 : Getting a Difference in the utilization of the Total Energy**
* **Query 6 : INNER JOIN, between the two tables**
* **Query 7 : Subquery implemented resulting into finding out those countries having the production rate greater than 0 in year 2010**

### QUERY 1

```python
tic = time.time()
TotalConsProd= pd.read_sql('''SELECT c.Country, c.Y1990 as 'Consumption 1990',p.Y1990 as'Production 1990' , 
                                                c.Y2000 as 'Consumption 2000',p.Y2000 as'Production 2000' ,
                                                c.Y2005 as 'Consumption 2005',p.Y2005 as'Production 2005' ,
                                                c.Y2010 as 'Consumption 2010',p.Y2010 as'Production 2010' ,
                                                c.Y2011 as 'Consumption 2011',p.Y2011 as'Production 2011' ,
                                                c.Y2012 as 'Consumption 2012',p.Y2012 as'Production 2012' ,
                                                c.Y2013 as 'Consumption 2013',p.Y2013 as'Production 2013' ,
                                                c.Y2014 as 'Consumption 2014',p.Y2014 as'Production 2014'
                            FROM TotalCons c,TotalProd p
                            WHERE c.Country = p.Country;''', conn)
toc = time.time()
tq1 = toc-tic
print ("Query run time:", tq1)
TotalConsProd
```

### QUERY 2

```python
tic = time.time()
TotalConsAVG= pd.read_sql_query('SELECT Country, SUM(Y1990+Y2000+Y2005+Y2010+Y2011+Y2012+Y2013+Y2014)/9 as `Total_Energy_Avg_Cons`'
                       'FROM TotalCons '
                       'GROUP BY Country '
                       'ORDER BY -Total_Energy_Avg_Cons', conn)
toc = time.time()
tq2 = toc-tic
print ("Query run time:", tq2)
TotalConsAVG
```

### QUERY 3

```python
tic = time.time()
TotalCons_Prod1990= pd.read_sql_query('''SELECT TotalCons.Country,TotalCons.Y1990 AS 'Consumption1990',TotalProd.Y1990 AS'Production1990' 
                                      FROM TotalCons, TotalProd 
                                      WHERE TotalCons.Country=TotalProd.Country 
                                      AND TotalCons.Y1990 > TotalProd.Y1990 
                                      Group BY TotalCons.Country''', conn)
toc = time.time()
tq3 = toc-tic
print ("Query run time:", tq3)
TotalCons_Prod1990
```

### QUERY 4

```python
tic = time.time()
TotalCons_Prod1990= pd.read_sql_query('''SELECT TotalCons.Country,TotalCons.Y1990 AS 'Consumption1990',TotalProd.Y1990 AS'Production1990' 
                                      FROM TotalCons, TotalProd 
                                      WHERE TotalCons.Country=TotalProd.Country 
                                      AND TotalCons.Y1990 > TotalProd.Y1990 
                                      Group BY TotalCons.Country 
                                      ORDER BY TotalProd.Y1990 DESC''', conn)
toc = time.time()
tq4a = toc-tic
print ("Query run time:", tq4a)
TotalCons_Prod1990
```

```python
tic = time.time()
TotalCons_Prod1990= pd.read_sql_query('''SELECT TotalCons.Country,TotalCons.Y1990 AS 'Consumption1990',TotalProd.Y1990 AS'Production1990' 
                                       FROM TotalCons, TotalProd 
                                       WHERE TotalCons.Country=TotalProd.Country 
                                       AND TotalCons.Y1990 < TotalProd.Y1990 
                                       Group BY TotalCons.Country''', conn)
toc = time.time()
tq4b = toc-tic
print ("Query run time:", tq4b)
TotalCons_Prod1990
```

### QUERY 5

```python
tic = time.time()
TotalCons_Prod1990= pd.read_sql_query('''SELECT TotalCons.Country,TotalCons.Y1990 AS 'Consumption 1990',TotalProd.Y1990 AS'Production 1990',
                                                                  TotalCons.Y1990-TotalProd.Y1990 AS 'Diffrence in 1990' 
                                        FROM TotalCons, TotalProd 
                                        WHERE TotalCons.Country=TotalProd.Country 
                                        AND TotalCons.Y1990 > TotalProd.Y1990 
                                        Group BY TotalCons.Country 
                                        LIMIT 30''', conn)
toc = time.time()
tq5 = toc-tic
print ("Query run time:", tq5)
TotalCons_Prod1990
```

### QUERY 6

```python
tic = time.time()
x= pd.read_sql_query('''SELECT c.Country,c.Y2010 ,p.Y2010 
                                      FROM TotalCons c
                                      INNER JOIN TotalProd p
                                      ON c.Country = p.Country''', conn)
toc = time.time()
tq6 = toc-tic
print ("Query run time:", tq6)
x
```

### QUERY 7

```python
tic = time.time()
y= pd.read_sql_query('''SELECT Country
                        FROM TotalCons
                        WHERE Y2010 IN (SELECT Y2010 
                                        FROM TotalProd
                                        WHERE Y2010 >0)''', conn)
toc = time.time()
tq7 = toc-tic
print ("Query run time:", tq7)
y
```
